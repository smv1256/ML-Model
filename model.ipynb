```python
# get data
import pandas as pd
df = pd.read_csv("socioeconomic_voting.csv")
df
```

```python
# data prep
y = df["County Median Household Income (2021)"]
y
x = df.drop(["County Median Household Income (2021)", "Party", "State", "County Name"], axis=1)
x
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
x_test
x_train
```

```python
# build model - random forest

# train
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=1)
rf.fit(x_train, y_train)

y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
# display data
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

rf_results = pd.DataFrame(["Random Forest", rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ["Method", "Training Mean Squared Error", "Training r^2", "Test Mean Squared Error", "Test r^2"]
rf_results
```
